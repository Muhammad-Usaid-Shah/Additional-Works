```python
import csv

# open the file
f = open('Assignment 1.csv', 'r')
r = csv.reader(f)
next(r)  
```

```python
c_list = []  
```

```python
for row in r:
    val = float(row[1])  
    c_list.append(val)
```

```python
f.close()
```

```python
# calculations
total_c = sum(c_list)
avg_c = total_c / len(c_list)
max_c = max(c_list)
min_c = min(c_list)
```

```python
# convert to kWh
c_kwh = [x / 1000 for x in c_list]
```

```python
# print results
print("Total:", total_c)
print("Average:", avg_c)
print("Max:", max_c)
print("Min:", min_c)
print("First 10 kWh values:", c_kwh[:10])
```
