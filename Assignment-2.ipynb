Step-1: Load CSV manually and store columns

```python
import csv

# open the file
f = open('Assignment 2.csv', 'r')
r = csv.reader(f)
next(r)  
```

```python
days = []
hours = []
consumptions = []
```

```python
# store each column in its own list
for row in r:
    days.append(int(row[0]))          
    hours.append(int(row[1]))         
    consumptions.append(float(row[2]))  

f.close()
```

Step-2: Group consumption by hour

```python
hour_dict = {}

for i in range(len(hours)):
    h = hours[i]
    c = consumptions[i]
    if h not in hour_dict:
        hour_dict[h] = []
    hour_dict[h].append(c)
```

```python
# Step 2.1 & 2.2: Compute averages and peaks
hour_avg = {}
hour_peak = {}
```

```python
for h, vals in hour_dict.items():
    hour_avg[h] = sum(vals) / len(vals)
    hour_peak[h] = max(vals)
```

```python
# Step 2.3: Print
print("Hour-wise consumption:")
for h in sorted(hour_dict.keys()):
    print(f"Hour {h}: Avg={hour_avg[h]:.2f}, Peak={hour_peak[h]}")
```

Step 3: Group consumption by day

```python
day_dict = {}
```

```python
for i in range(len(days)):
    d = days[i]
    c = consumptions[i]
    if d not in day_dict:
        day_dict[d] = []
    day_dict[d].append(c)
```

```python
# Step 3.2 & 3.3: Total and daily average
day_total = {}
day_avg = {}
```

```python
for d, vals in day_dict.items():
    day_total[d] = sum(vals)
    day_avg[d] = sum(vals) / len(vals)
```

Results

```python
print("\nDay-wise consumption:")
for d in sorted(day_dict.keys()):
    print(f"Day {d}: Total={day_total[d]:.2f}, Avg={day_avg[d]:.2f}")
```
