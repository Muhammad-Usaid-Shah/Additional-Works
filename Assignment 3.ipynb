```python
import csv

# 1. Load both CSVs

# Load first CSV (loads)

f1 = open('Assignment 3 load.csv', 'r')
r1 = csv.reader(f1)
next(r1) 
```

```python
loads = []
for row in r1:
    loads.append(float(row[1])) 
f1.close()
```

```python
# Load second CSV (prices)

f2 = open('Assignment 3 price.csv', 'r')
r2 = csv.reader(f2)
next(r2)
```

```python
prices = []
for row in r2:
    prices.append(float(row[1]))  
f2.close()
```

```python

# 3. Write a function to compute

def compute_bill(loads, prices):
    
    # 3.1 Hourly bill
    
    hourly_bills = []
    for i in range(len(loads)):
        hourly_bills.append(loads[i] * prices[i])
        
    # 3.2 Total bill
    
    total_bill = sum(hourly_bills)
    return hourly_bills, total_bill
```

```python
# Use the function

hourly, total = compute_bill(loads, prices)
```

```python
# Print results

print("Hourly bills (first 10):", hourly[:10])
print("Total bill:", total)
```
